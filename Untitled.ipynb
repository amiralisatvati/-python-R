```python
import pandas as pd
import requests

new_ulr = "https://api.themoviedb.org/3/discover/movie?api_key="
api_key = "2a196374a6d1fa85a77f6c7883e7bf5e"
query = "&primary_release_date.gte=2022-01-01&primary_release_date.lte=2023-02-29&with_genres=28" # Assuming you want action movies, with_genres=28
final_url = new_ulr + api_key + query

data = requests.get(final_url).json()

# Check if 'results' key exists in data
if 'results' in data:
    df = pd.DataFrame(data['results'])  # Create DataFrame from 'results'
else:
    # Handle the case where 'results' key is not found
    print("Error: 'results' key not found in API response.")
    print("Data structure:", data)
    # This will help in understanding the API response structure
    # You can then modify the code to access the correct key containing the data you need
```

```python
movie_id = [
    0, 299534, 19995, 140607, 299536, 597, 135397,
    420818, 24428, 168259, 99861, 284054, 12445, 181808,
    330457, 351286, 109445, 321612, 260513
]
```

```python
basic_url = 'https://api.themoviedb.org/3/movie/{}?api_key={}'
```

```python
json_list = []
for movie in movie_id:
    url = basic_url.format(movie, api_key)
    r = requests.get(url)
    if r.status_code != 200:
        continue
    else:
        data = r.json()
        json_list.append(data)
df = pd.DataFrame(json_list)
```

```python
df=df.loc[:,["title","id","revenue"]].sort_values(by= "revenue", ascending  =False)
```

```python
df.to_json("movie.json")
```
